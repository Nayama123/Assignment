## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
import pandas as pd

# Load the dataset
df = pd.read_csv(r"C:\Users\grace\Downloads\hour.csv")  

# Step 1: Choose a column
col = 'cnt'

# Step 2: Calculate IQR
Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

# Step 3: Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Step 4: Identify outliers
outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

# Display the result
print(f"Total outliers in '{col}':", outliers.shape[0])
outliers[[col, 'dteday', 'hr']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv(r"C:\Users\grace\Downloads\hour.csv")

# Set style
sns.set(style="whitegrid")

# Plot
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='cnt', color='skyblue')
plt.title("Boxplot of Total Bike Rentals (`cnt`)")
plt.xlabel("Total Rentals")
plt.show()
```

cnt – This column represents the total number of bike rentals per hour, combining both casual and registered users.

Approach:
We used the Interquartile Range (IQR) method to detect outliers.

The IQR is the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data.

Outliers are defined as values:

Below Q1 − 1.5 × IQR, or

Above Q3 + 1.5 × IQR

 Result:
Total number of outliers in cnt: 505

 Interpretation:
These outliers may indicate unusually high or low bike usage, possibly due to:

Public holidays

Bad weather or events

Data collection anomalies

##Q.14 Find the correlation between numerical columns and discuss any interesting relationships.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\grace\Downloads\hour.csv")


# Compute correlation matrix for numerical columns
correlation_matrix = df.corr(numeric_only=True)

# Display heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```


| Variable Pair         | Correlation | Insight |
|-----------------------|-------------|---------|
| **`cnt` vs `registered`** | **0.97** | Total rentals are strongly driven by registered users |
| **`cnt` vs `casual`**     | **0.69** | Moderate to strong influence from casual users |
| **`temp` vs `cnt`**       | **0.40** | Warmer temperatures encourage more bike rentals |
| **`hum` vs `cnt`**        | **-0.32** | High humidity discourages biking |
| **`windspeed` vs `cnt`** | **-0.09** | Slight negative impact from windspeed |
| **`temp` vs `atemp`**     | **0.99** | Very strong correlation between actual and perceived temperature |

 Interpretation:
- Rentals (`cnt`) are **more influenced by registered users** than casual users.
- **Weather conditions** (temperature, humidity, windspeed) play a moderate role in bike rental volume.
- This analysis is crucial for predicting demand and planning resource allocation (e.g., station balancing).
